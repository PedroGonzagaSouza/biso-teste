```python
import numpy as np 
import pandas as pd
import sqlite3
```

```python
# filmes = pd.read_csv('./data/filmes.csv', sep=',')
# ratings = pd.read_csv('./data/ratings.csv', sep=',')
filmesList = []
ratingsList = []    
conn = sqlite3.connect('filmes.db')
cur = conn.cursor()
for row in cur.execute('SELECT * FROM FILMES'):
    filmesList.append(row)
    
for row in cur.execute('SELECT * FROM RATINGS'):
    ratingsList.append(row)
conn.commit()
conn.close()    
ratingsdf = pd.DataFrame(ratingsList, columns=['userId', 'movieId', 'rating'])
filmesdf = pd.DataFrame(filmesList, columns=['movieId', 'title', 'genres', 'year'])



```

```python
df = filmesdf.merge(ratingsdf, on='movieId')
df.head()
```

```python
tabelaFilmes = df.pivot_table(index='title', columns='userId', values='rating').fillna(0)
tabelaFilmes.head()
```

```python
import sklearn.metrics.pairwise as pw
```

```python
recomendacoes = pw.cosine_similarity(tabelaFilmes)
recomendacoes = pd.DataFrame(recomendacoes, index=tabelaFilmes.index, columns=tabelaFilmes.index)
recomendacoes[''].sort_values(ascending=False).head(10)
```
